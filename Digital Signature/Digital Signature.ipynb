# Digital Signature

- A digital signature is an electronic equivalent of a handwritten signature.
- It ensures **authenticity**, **integrity** and **non-repudiation** of messages.
- Based on **public key cryptography**

### Process

1. Sender generates a hash of the message.
2. Hash is encrypted with sender's private key.
3. Receiver verifies using sender's public key.

### Properties

- **Authenticity**: Confirms the sender's identity.
- **Integrity**: Detects message alteration.
- **Non-repudiation**: Prevents sender from denying authorship.
- **Unforgeability**: Cannot be copied or imitated.

## RSA Digital Signature

The RSA algorithm can also be used to create **digital signatures**, which verify the authenticity and integrity of a message.

- The sender uses their *private key* to sign their message.
- The receiver verifies the signature using the sender's *public key*.

### Algorithm:

1. The sender computes a message digest (hash) of the original message:

$$h = H(M)$$

2. The sender encrypts the hash with their private key to create the digital signature:

$$S=h^d\text{ }mod\text{ }n$$

3. The signature is sent along with the original message: $(M,S)$

### Verification:

1. Receiver computes the message hash:

$$h_1 = H(M)$$

2. Receiver decrypts the received signature using the sender's public key:

$$h_2=S^e\text{ }mod\text{ }n$$

3. If $h_1 = h_2$ the signature is valid, otherwise, the message or signature has been altered.

## Implementation

```python
# Generate private and public key of sender:
from Crypto.PublicKey import RSA

key = RSA.generate(2048)
private_key = key.export_key()
with open("private.pem", "wb") as f:
    f.write(private_key)
public_key = key.publickey().export_key()
with open("public.pem", "wb") as f:
    f.write(public_key)
print("Keys generated and saved to 'private.pem' and 'public.pem'")
```

```python
# Hashing the message
from Crypto.Hash import SHA256
message = b'This is a secret message.'
hash_obj = SHA256.new(message)
print("Original Message:", message.decode())
print("Message Hash:", hash_obj.hexdigest())
```

```python
# Senders signs the hash with his private key
from Crypto.Signature import pkcs1_15
private_key = RSA.import_key(open("private.pem").read())
signature = pkcs1_15.new(private_key).sign(hash_obj)
print("Digital Signature:", signature.hex())

print("Sent Object: ("+ message.decode() + ", " + signature.hex() + ")")
```

```python
# Verifier verifies the signature with sender's public key
from Crypto.Signature import pkcs1_15
public_key = RSA.import_key(open("public.pem").read())
h1 = SHA256.new(message)
try:
    pkcs1_15.new(public_key).verify(h1, signature)
    print("The signature is valid. The message is authentic.")
except (ValueError, TypeError):
    print("The signature is not valid. The message may have been tampered with.")
```
