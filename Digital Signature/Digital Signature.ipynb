# Digital Signature

- A digital signature is an electronic equivalent of a handwritten signature.
- It ensures **authenticity**, **integrity** and **non-repudiation** of messages.
- Based on **public key cryptography**

## Process

1. Sender generates a hash of the message.
2. Hash is encrypted with sender's private key.
3. Receiver verifies using sender's public key.

## Properties

- **Authenticity**: Confirms the sender's identity.
- **Integrity**: Detects message alteration.
- **Non-repudiation**: Prevents sender from denying authorship.
- **Unforgeability**: Cannot be copied or imitated.
