# ElGamal Algorithm

ElGamal Algorithm is a public key cryptosystem based on **Discrete Logarithm Problem**.

## Key Generation:

1. Select a large prime number $p$ and a generator $g$ of the multiplicative group $Z^*_p$.
2. Select a private key such that $1\le x \le p-2$
3. Compute $h = g^x\texttt{ }mod\texttt{ }p$. The public key is (p, g, h) and the private key is $x$.

## Encryption

To encrypt a message M:
- Choose a random integer $k$ such that $1 \le k \le p−2$.
- Compute $C_1 = g^k\texttt{ }mod\texttt{ }p$
- Compute $C_2 = M\cdot h^k\texttt{ }mod\texttt{ }p$
- The Ciphertext is $(C_1, C_2)$

## Decryption

To decrypt the ciphertext $(C_1,C_2)$ using the private key $x$:
- Compute the shared secret $s = C^x_1\texttt{ }mod\texttt{ }p$
- Compute the $s^{-1}\texttt{ }mod\texttt{ }p$ i.e the modular inverse of s.
- Compute the original message $M = C_2\cdot s^{-1}\texttt{ }mod\texttt{ }p$

## Verification

The ElGamal decryption expression $C_2 \cdot (C_1^d)^{-1}$ can be verified to be $P$ through substitution:
$$[C_2 \cdot (C_1^d)^{-1}]\texttt{ }mod\texttt{ }p = [(e_2^{r} \cdot P)\cdot (e_1^{rd})^{-1}]\texttt{ }mod\texttt{ }p$$
$$i.e$$
$$\boxed{(e_1^{dr})\cdot P \cdot (e_1^{rd})^{-1} = P}$$

$\boxed{Note}$ For the ElGamal cryptosystem to be secure, p must be at least 300 digits and r must be new for each encipherment. 

```python
import sympy
import random
```

```python
# Key Generation
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def generate_keys(p, g):
    # Generate large prime number p
    p = sympy.randprime(10**5, 10**6)

    g = 5
    # Generate generator g
    while gcd(g, p - 1) != 1:
        g = sympy.randprime(2, p - 1)

    return p, g

def select_private_key(p):
    return random.randint(1, p - 2)

def compute_public_key(g, private_key, p):
    return g**private_key % p

p, g = generate_keys(0, 0)
private_key = select_private_key(p)
public_key = compute_public_key(g, private_key, p)

with open(".env", "w") as f:
    f.write("P=" + str(p) + "\n")
    f.write("G=" + str(g) + "\n")
    f.write("PRIVATE_KEY=" + str(private_key) + "\n")
    f.write("PUBLIC_KEY=" + str(public_key) + "\n")

print("Prime (p):", p)
print("Generator (g):", g)
print("Public Key:", public_key)
print("Private Key:", private_key)
```

```python
# Encryption
def encrypt(message, public_key, g, p):
    # Generate random integer k
    k = random.randint(1, p - 2)

    # Compute c1 and c2
    c1 = g**k % p
    c2 = (message * public_key**k) % p

    return c1, c2, k

Message = "Hi! This is Elgamal implementation in Python"
Message_arr = [ord(char) for char in Message]
cipher_arr = []
for i in range(len(Message_arr)):
    c1, c2, k = encrypt(Message_arr[i], public_key, g, p)
    Cipher = (c1, c2)
    cipher_arr.append(Cipher)
print("Ciphertext (c1, c2):", cipher_arr)
print("Random k:", k)
```

```python
# Decryption
def decrypt(ciphertext, private_key, p):
    c1, c2 = ciphertext
    s = c1**private_key % p
    s_inv = pow(s, -1, p)
    message = (c2 * s_inv) % p
    return message

decrypted_message = []
for cipher in cipher_arr:
    decrypted_message.append(decrypt(cipher, private_key, p))
decrypted_message = ''.join(chr(num) for num in decrypted_message)
print("Decrypted Message:", decrypted_message)
```
