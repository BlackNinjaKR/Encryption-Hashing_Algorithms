# Additive Cipher / Shift Cipher / Caesar Cipher

The simplest mono-alphabetic substitution cipher is additive cipher. As the name suggests, addition modulus 26 operation is performed on the plain-text to obtain a cipher-text.
$$C = (M + k)\texttt{ }mod\texttt{ }26 \\ M = (C - k)\texttt{ }mod\texttt{ }26$$
where,  
- $C$ -> cipher-text  
- $M$ -> message/plain-text  
- $k$ -> key

The key space is 26.

```python
k = [-100, -15, -3, 10, 20, 30]
msg = "Hello, World!"

def encrypt(msg, k):
    cipher = ""
    k = k % 26
    
    for c in msg:
        if c.isupper():
            cipher += chr((ord(c) - ord('A') + k) % 26 + ord('A'))
        elif c.islower():
            cipher += chr((ord(c) - ord('a') + k) % 26 + ord('a'))
        else:
            cipher += c
    
    return cipher


def decrypt(cipher, k):
    return encrypt(cipher, -k)
```

```python
for key in k:
    cipher = encrypt(msg, key)
    print("Cipher:", cipher)
    
    decrypted_msg = decrypt(cipher, key)
    print("Decrypted:", decrypted_msg)
    print()
```
